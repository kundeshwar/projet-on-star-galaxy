```python
#making scatter plot of first 12 star by using there RA and declination

# QUESTION 1 (PART 1)
import matplotlib.pyplot as mp
import csv
import math
with open("C:\\Users\\kunde\\Downloads\\stardata (1)2.csv") as csvfile:
    a = csv.DictReader(csvfile)
    
    count=0
    apparent_magnitude = []
    parallax = []
    
    for column in a:
        count = count + 1
        apparent_magnitude.append(column["APPARANT MAGNITUDE"])
        parallax.append(column["PARALLAX"])
        if count > 201:
            break
apparent_magnitude2 = []
for i in range(len(parallax)):
    apparent_magnitude2.append(float(apparent_magnitude[i]))
    if i == 201:
        break            
            
a = []
            
for i in range(len(parallax)):
    a.append(float(parallax[i]))
    if i == 201:
        break

distance = [ 1/i for i in a]
log10_d = [(5* math.log10(j)) for j in distance]
# -5+5*log10_d=c

c = [ (-5+i) for i in log10_d ]

absolute_magnitude = [(apparent_magnitude2[i]-c[i]) for i in range(len(parallax))]
print(absolute_magnitude)

mp.hist(absolute_magnitude,edgecolor="black",bins=50)
mp.title("HISTOGRAM OF ABSOLUTE MAGNITUDE",fontsize=15)
mp.xlabel("VALUES OF ABSOLUTE MAGNITUDE",fontsize=15)


mp.show()
#PART 2(QUESTION 1)
import matplotlib.pyplot as mp
import csv
with open("C:\\Users\\kunde\\Downloads\\stardata (1).csv") as csvfile:
    a = csv.DictReader(csvfile)
    
    count=0
    ra = []
    declination = []
    apparent_magnitude = []
    
    for column in a:
        count = count + 1
        ra.append(column["RA(INDEGREE)"])
        declination.append(column["DECLINATION"])
        apparent_magnitude.append(round(float(column["APPARANT MAGNITUDE"]),0))
        if count > 12:
            break
sizes = [i*30 for i in apparent_magnitude]          

mp.xlabel("DECLINATION (IN DEGREES)",fontsize=15)
mp.ylabel("RA(IN DEGREES)",fontsize=15)
mp.title("STAR CLUSTER",fontsize=15)            
            
mp.scatter(ra,declination,edgecolor="r",c="b",linewidth=1,marker="*",s=sizes)
mp.show()

```
