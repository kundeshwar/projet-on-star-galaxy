```python
import matplotlib.pyplot as mp
#import csv
import pandas as pd
#var = pd.read_csv("C:\\Users\\kunde\\Downloads\\data.projet_3.csv")


#var.dropna().to_csv("kundeshwar3.csv", index=False)





with open("C:\\Users\\kunde\\Music\\kundeshwar3.csv")as csvfile:
    a = csv.reader(csvfile)


    count = 0
    distance_modulus = []
    velocity = []
    for i in a:
        count = count + 1
        distance_modulus.append(float(i[0]))
        velocity.append(float(i[1]))
        if count>4063:
            break
            
distance1 = [((i/5)-5) for i in distance_modulus]

distance2 = [(10**j) for j in distance1]#distance2 is distance in Mpc
print(distance2)

mp.title("SCATER PLOT",fontsize=15)
mp.ylabel("DISTANCE IN Mpc",fontsize=15)
mp.xlabel("VELOCITY IN m/s",fontsize=15)

mp.scatter(velocity,distance2)
mp.show()



  
```
